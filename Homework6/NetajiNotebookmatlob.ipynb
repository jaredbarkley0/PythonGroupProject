```python
# Import libraries
import pandas as pd
```

```python
# Loading in project data for use in Homework 6

compiledcountydata_dir = "https://raw.githubusercontent.com/jaredbarkley0/PythonGroupProject/master/MN%20County%20Data_032021.csv"
compiledcountydata = pd.read_csv(compiledcountydata_dir)
compiledcountydata.head()
```

### MatplotLib
#### Bar Graph

```python
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
# Total Positive Cases County wise in Bar Graph
plt.figure(figsize=(20,20))
plt.bar(compiledcountydata["County"], compiledcountydata["Total Positive Cases"])
plt.xticks(compiledcountydata["County"])
plt.ylabel('Total Positive Cases')
plt.xlabel('Country')
plt.xticks(compiledcountydata["County"], rotation='vertical', size=10)
plt.show()
```

```python
# Deaths County wise in Line Graph
plt.figure(figsize=(20,20))
plt.plot(compiledcountydata["County"], compiledcountydata["Total Deaths"])
plt.xticks(compiledcountydata["County"])
plt.ylabel('Total Deaths')
plt.xlabel('Country')
plt.xticks(compiledcountydata["County"], rotation='vertical', size=10)
plt.grid()
plt.show()

```

```python
# Total Postive Cases , Total Deaths by County in Stocked bar Graph

import numpy as np 
x = np.arange(87) 
plt.figure(figsize=(20,20))
width = 0.35  # the width of the bars
plt.bar(compiledcountydata["County"], compiledcountydata["Total Positive Cases"])
plt.bar(compiledcountydata["County"], compiledcountydata["Total Deaths"], color="r")
plt.xticks(compiledcountydata["County"])
plt.ylabel('Total Positive Cases')
plt.xlabel('Country')
plt.xticks(compiledcountydata["County"], rotation='vertical', size=10)
plt.show()
```

```python
## Total Postive Cases , Total Deaths by County by Bar and Line Graph

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(compiledcountydata["County"], compiledcountydata["Total Positive Cases"], color='g')
ax2.plot(compiledcountydata["County"], compiledcountydata["Total Deaths"], color='b')

ax1.set_xlabel('Total Postive Cases')
ax1.set_ylabel('County', color='g')
ax2.set_ylabel('Total Deaths', color='b')
ax1.set_xticklabels(compiledcountydata["County"], rotation='vertical', size=10)


fig.show()


```
