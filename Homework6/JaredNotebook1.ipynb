# Matplotlib

```python
# import matplotlib and inline magic 

%matplotlib inline

import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
# basic plot of number of cases vs pop density M.Noonan

fig, ax = plt.subplots()

casesbydensityscatter = ax.scatter(compiledcountydata['Pop Density'], compiledcountydata['Total Positive Cases'], 
                                   c=compiledcountydata['Total Deaths'], label=compiledcountydata['Total Deaths'])
ax.set(title = 'Covid Cases per Population/Sq Mile', xlabel = 'Population/Sq Mi', ylabel='Total Cases')
#Setting the size rather large to that is fills screen
fig.set_size_inches(15,12)

totaldeathlegend = ax.legend(*casesbydensityscatter.legend_elements(),loc='upper left',title = "Deaths")

ax.add_artist(totaldeathlegend)

# plt.plot(compiledcountydata['Pop Density'],compiledcountydata['Total Positive Cases'])
plt.show
```
