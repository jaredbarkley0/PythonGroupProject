```python
facilitydataraw = pd.read_csv(countyfacilities_dir)
facilitydataraw
```

```python
# store10764sorted = (store10764labordata[['Date','Manager','Projected Sales','Sales','DT TTL','Day']]).groupby(['Manager','Date','Day'])
# dailysalereport = store10764sorted['Sales'].agg(DailySales = 'sum', MaxSales = 'max')
# dailysalereport

#Create a Groupby Object for Raw Facilities Data

facilitygrouper = facilitydataraw[['Name of Facility','County','Total confirmed cases in facility']].groupby('County')
countyfacilitytally = facilitygrouper['Name of Facility'].agg(FacilitiesinCounty = 'count')
                                       
countyfacilitytally
```

```python
countyfacilitycases = facilitygrouper['Total confirmed cases in facility'].agg(Casesinfacility = 'sum')
countyfacilitycases
```

```python
countyfacilitycases.loc['Hennepin']
```

```python
compiledfacilitydata = pd.concat([countyfacilitycases,countyfacilitytally], axis=1)
#compiledcountydata['Compare County'] = compiledcountydata.index
compiledfacilitydata
```

```python
len(countycasesraw['County'])
```

```python
len(compiledfacilitydata.index)
```

```python
finalcountydata = countycasesraw.merge(compiledfacilitydata, left_on='County', right_on='County', how = 'outer')
finalcountydata
```

```python
finalcountydata.to_csv("MN County Data.csv")
```
